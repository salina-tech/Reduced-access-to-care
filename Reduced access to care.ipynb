```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
labels = ['Round1', 'Round2', 'Round3']
eighteen_Fourty_Four = [2.4, 3.2, 3.8]
Fourty_Five_Sixty_Four = [3.3, 3.3, 2.6]
Sixty_Five_Over = [4.3, 3.7, 2.7]

x = np.arange(len(labels))  # the label locations
width =0.09# the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x-0.1, eighteen_Fourty_Four, width, label='18 to 44')
rects2 = ax.bar(x, Fourty_Five_Sixty_Four, width, label='45 to 64')
rects3 = ax.bar(x+0.1, Sixty_Five_Over, width, label='65+')

plt.xlabel('Round') 
Round = [1,2,3]

ax.set_ylabel('Percentage of reduced access to care')
ax.set_title('Percentage of reduced access to care for any other reason than COVID-19')
ax.set_xticks(x, labels)
ax.legend()

stacked=False
plt.xticks(x, ['Round1', 'Round 2', 'Round 3'])


fig.tight_layout()


plt.show()
```
